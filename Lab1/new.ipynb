```python
import time
import random
import numpy as np


def set_cover(universe, subsets, costs):
    cost = 0
    # Check if all the numbers in the subsets can achieve the goal
    elements = set(e for s in subsets for e in s)
    if elements != universe:
        return None

    covered = set()
    cover = []

    n=0

    # Do the while cycle until we have found the right subset
    while covered != elements:
        n += 1
        # subset = setwith the highest 'elements not in covered/ cost' ratio
        if(len(covered) == 0):
            subset = max(subsets, key = lambda s: len(s))
        else:
            subset = max(subsets, key = lambda s: len(s - covered)/costs[subsets.index(s)])
        #subset = max(subsets, key = lambda s: len(s - covered) * costs[subsets.index(s)])
        
        #print("subset")
        #print(subset)

        #print(subset - covered)

        # cover + subset
        cover.append(subset)
        
        #print("cover")
        #print(cover)

        # cost + cost of subset
        cost += costs[subsets.index(subset)]
        
        #print("cost")
        #print(cost)

        # covered + elements in subset not in covered
        covered |= subset
        
        #print("Covered")
        #print(covered)
 
    return cover, cost, n
```

```python
def main(a, b , c, x=time.time()):
    m = a
    universe = set(range(0, m))
    sub = b  
    
    subsets = [set(x) for x in sub]
    costs =  c 
    cover = set_cover(universe, subsets, costs)

    print('covering sets = ', cover[0])
    print('weigth = ', cover[1])
    print('time: ', time.time()-x)
    print('nodes: ', cover[2])
     
```

```python
def problem(N, seed=None):
    random.seed(seed)
    return [
        list(set(random.randint(0, N - 1) for n in range(random.randint(N // 5, N // 2))))
        for n in range(random.randint(N, N * 5))
    ]
```

```python
# Il costo si deve basare su quanti numeri mancano al completamento e alla lunghezza del subset
# a pari valore sui numeri, il subset piu lungo deve avere un costo maggiore

m1 = 500
S1 = problem(m1, 42)
#S1 = [[1], [1,3], [1,2], [3], [1,4]]

#print(S1)

P1 = []
for vec in S1:
    P1.append(len(vec))

#print(P1)

#P1 = [2,1,1,2,1]

if __name__ == '__main__':
    main(m1,S1,P1)
```
